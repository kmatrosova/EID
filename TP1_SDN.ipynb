```python
# A - Importation des libraires

from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
# B - Manipulation d’un jeu de données

iris = datasets.load_iris()

print("\033[1m Données: \033[0m \n", iris.data, " \n")
print("\033[1m Variables: \033[0m \n", iris.feature_names, " \n")
print("\033[1m Classes: \033[0m \n", iris.target_names, " \n")

print("\033[1m Nom des classes pour chaque donnée: \033[0m")
for i in range (0, len(iris.target)):
    print(" X", i, " : ", iris.target_names[(iris.target[i])])


print("\n")

print("\033[1m Moyenne: \033[0m \n", iris.data.mean(0))
print("\033[1m Écart-type: \033[0m \n", iris.data.std(0))
print("\033[1m Minimum: \033[0m \n", iris.data.min(0))
print("\033[1m Maximum: \033[0m \n", iris.data.max(0), " \n")

print("\033[1m Nombre de données: \033[0m", iris.data.shape[0])
print("\033[1m Nombre de variables: \033[0m", iris.data.shape[1])
print("\033[1m Nombre de classes: \033[0m", iris.target_names.size, " \n")

```

```python
# C - Téléchargement et importation de données

mnist = loadmat("mnist-original.mat")
mnist_data = mnist["data"].T
mnist_label = mnist["label"][0]

print("\033[1m Variables: \033[0m \n", mnist_data, " \n")
print("\033[1m Nombre de données: \033[0m ", mnist_data.shape[0], " \n")
print("\033[1m Nombre de variables: \033[0m ", mnist_data.shape[1], " \n")

print("\033[1m Numéros des classes pour chaque donnée: \033[0m", mnist_label, " \n")

print("\033[1m Moyenne: \033[0m \n", mnist_data.mean(0), " \n")
print("\033[1m Écart-type: \033[0m \n", mnist_data.std(0), " \n")
print("\033[1m Minimum: \033[0m \n", mnist_data.min(0), " \n")
print("\033[1m Maximum: \033[0m \n", mnist_data.max(0), " \n\n")

print("\033[1m Nombre de classes: \033[0m ", len(np.unique(mnist_label)))
```

```python
# D - Génération de données et affichage

colors=['green','skyblue', 'navy', 'cyan', 'blue']

X, Y = datasets.make_blobs(n_samples=1000, centers=4)

plt.figure()
for i in range(0, X.shape[0]):
      plt.scatter(X[i][0], X[i][1], c=colors[Y[i]%len(colors)])

plt.title("Mes données")
plt.xlim(left=-15, right=15)
plt.ylim(bottom=-15, top=15)
plt.xlabel("axe des abscisses")
plt.ylabel("axe des ordonnées")
plt.show()
plt.close()



X1, Y1 = datasets.make_blobs(n_samples=100, centers=2)
X2, Y2 = datasets.make_blobs(n_samples=500, centers=3)

X3 = np.vstack((X1, X2))
Y3 = np.hstack((Y1, Y2))

plt.figure()

for i in range(0, X3.shape[0]):
      plt.scatter(X3[i][0], X3[i][1], c=colors[Y3[i]%len(colors)])

plt.show()
plt.close()
```
